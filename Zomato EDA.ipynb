```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
# combing the dataset
```

```python
# It is similar to joins
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(5)
```

```python
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## Pie Chart - Top 3 Countries that uses Zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

# Observation : India has the most records or transactions on Zomato followed by United States and United Kingdom

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
```

```python
# changing the column name of the count 0 > Rating Count
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When rating is between 4.5 to 4.9 -> Excellent
2. When rating is between 4.0 to 3.4 -> very good
3. When rating is between 3.5 to 3.9 -> good
4. When rating is between 3.0 to 3.4 -> average
5. When rating is between 2.5 to 2.9 -> average
6. When rating is between 2.0 to 2.4 -> poor

```python
ratings.head()
```

```python
# increasing the size of plot
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
# plotting according colors
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','yellow','green','green'])
```

## Observation:
    1. Not Rated count is very high 
    2. Maximum no of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data= ratings,palette=['blue','red','yellow','green','green'])
```
